```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2)
```

```python
print(f'Training value of data: {xtrain}')
print(f'Testing values of data: {xtest}')
print(f'Training labels of data: {ytrain}')
print(f'Testing labels of data: {ytest}')
print('\n')
print(f'Shape of Training value of data: {xtrain.shape}')
print(f'Shape of Testing values of data: {xtest.shape}')
print(f'Shape of Training labels of data: {ytrain.shape}')
print(f'Shape of Testing labels of data: {ytest.shape}')
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
prediction =  model.predict(xtest)
```

```python
print('Prediction :\n', prediction)
```

```python
print(model.score(x, y))
```
