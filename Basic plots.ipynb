```python
import matplotlib.pyplot as plt
```

```python
x = [i for i in range(10)]
y = [i*2 for i in range(10)]
print(x)
print(y)
```

```python
fig, ax = plt.subplots(figsize=(6,5))
plt.plot(x,y)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Regression Line")
```

```python
fig, ax = plt.subplots(figsize = (6,5))
plt.scatter(x,y)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
```
